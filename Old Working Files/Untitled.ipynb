```python
import PIL 
import numpy as np
import os
import nibabel as nib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_path = r'C:\Users\schmi\data\NII\\'
images = []
segmentations = []
for f in os.listdir(data_path):
  if '_K' in f:
    continue
  else:
    images.append(f)
    segmentations.append(f.replace('.nii', '_K.nii'))

print(images[0], segmentations[0])
images = np.array(images)
segmentations = np.array(segmentations)

indices = np.array(range(len(images))) # we will use this in the next step.
```

```python
path_img = (data_path+'\\'+images[0])
nifti_image = nib.load(path_img)
nifti_data_img = nifti_image.get_fdata()
nifti_img = nifti_data_img[:,:,50]
path_seg = (data_path+'\\'+segmentations[0])
nifti_seg= nib.load(path_seg)
nifti_data_seg = nifti_seg.get_fdata()
nifti_seg = nifti_data_seg[:,:,50]
```
